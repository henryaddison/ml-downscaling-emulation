```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cftime
```

```python
gcmdata = xr.open_mfdataset("../../../data/60km/rcp85/01/*/day/*.nc")
```

```python
gcm_single_day = gcmdata.sel(ensemble_member=1, time=cftime.Datetime360Day(1980, 12, 1, 12, 0, 0, 0))
```

```python
plt.figure(figsize=(14,6))

cp_model_rotated_pole = ccrs.RotatedPole(pole_longitude=177.5, pole_latitude=37.5)
platecarree = ccrs.PlateCarree()
ax = plt.axes(projection=platecarree)

gcm_single_day.psl.plot(ax=ax, x='longitude', y='latitude', add_colorbar=False, transform=platecarree)

ax.coastlines()
```

```python
plt.figure(figsize=(14,6))

ax = plt.axes(projection=platecarree)

gcm_single_day.pr.plot(ax=ax, x='longitude', y='latitude', add_colorbar=False, transform=platecarree)

ax.coastlines()
```
